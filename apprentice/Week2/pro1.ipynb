```python
import numpy as np

data = np.loadtxt('data/chocolate_rating.csv', delimiter=',')
print(data)
```

```python
rating_mean = data[:, 3].mean()
print(rating_mean)
```

```python
high_level = data[data[:, 3] >= 4]
print(high_level)
```

```python
high_id = data[:, 0].astype(int)
print(high_id)
```

```python
include_kakao = high_level[:, 2]
max_include = np.max(high_level[:, 2])

# 결과 출력
print("최대 함유량:", max_include)

frequency = np.count_nonzero(include_kakao == max_include)


print("빈도수:", frequency)
```
